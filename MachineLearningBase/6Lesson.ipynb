```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
plt.figure(figsize = (8,8))
area = [48,30,20,15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#7CB342','#C0CA33', '#FFB300','#F57C00']
plt.pie(area, labels = labels, colors = colors, startangle = 45, autopct = '%1.1f%%', shadow = 'true', explode = [0,0, 0,0.1])
plt.legend()
plt.show()
```

```python
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = 'midnightblue')
plt.gca().set(title = 'Histogram', ylabel = 'Freq')
plt.show()
```

```python
x = np.random.rand(2000)
plt.hist(x, bins = 30, color = 'red')
plt.gca().set(title = 'Histogram', ylabel = 'Freq')
plt.show()
```

```python
plt.figure(figsize = (10,8))
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = 'midnightblue', edgecolor = 'red')
plt.gca().set(title = 'Histogram', ylabel = 'Freq')
plt.show()
```
