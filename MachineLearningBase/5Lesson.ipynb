```python
plt.figure(figsize=(10,5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x,2)
y2 = np.power(x,3)
plt.plot(x,y1,'b-',label ='$y1 = x^2$')
plt.plot(x,y2,'r-',label ='$y1 = x^3$')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
x = np.linspace(0,10,2000)
plt.plot(x,np.sin(x),'b-',label ='$sin(X)')
plt.plot(x,np.cos(x),'r-',label ='$cos(X)$')
plt.xlabel(r'$X$',fontsize = 18)
plt.ylabel(r'$X$',fontsize = 18)
plt.title('$sin(X)$ $ & $ $cos(x)$',fontsize = 14)
plt.legend(loc = 'upper right')
plt.show()
```

```python
plt.figure(figsize=(10,6))
x = np.linspace(0,10,2000)
plt.plot(x,np.sin(x),'b-',label ='$sin(X)$')
plt.plot(x,np.cos(x),'r-',label ='$cos(X)$')
plt.fill_between(x, 0, np.sin(x))
plt.fill_between(x, 0 ,np.cos(x))
plt.xlabel(r'$X$',fontsize = 18)
plt.ylabel(r'$X$',fontsize = 18)
plt.title('$sin(X)$ $ & $ $cos(x)$',fontsize = 14)
plt.legend(loc = 'lower left')
plt.show()
```

```python
x1 = np.array([250,150,350,252,450,550,455,358,158,355])
y1 = np.array([40,50,80,90,100,50,60,88,54,45])
x2 = np.array([200,100,300,220,400,500,450,380,180,350])
y2 = np.array([400,500,800,900,1000,500,600,808,504,405])

plt.scatter(x1,y1)
plt.xlabel('$Time$ $ Spend$ $', fontsize = 12)
plt.ylabel('$Score $', fontsize = 12)
plt.title('Scatter Graph')
plt.show()
```

```python
plt.scatter(x1,y1,label= 'Class 1')
plt.scatter(x2,y2,label= 'Class 2', color = 'r')
plt.xlabel('$Time$ $ Spend$ $', fontsize = 12)
plt.ylabel('$Score $', fontsize = 12)
plt.title('Scatter Graph')
plt.show()
```

```python
plt.scatter(x1,y1,label= 'Class 1', marker = 'o', color = 'm')
plt.scatter(x2,y2,label= 'Class 2', color = 'r', marker = 'v')
plt.xlabel('$Time$ $ Spend$ $', fontsize = 12)
plt.ylabel('$Score $', fontsize = 12)
plt.title('Scatter Graph')
plt.show()
```

```python
plt.figure(figsize = (9,9)) 
area = [48,30,20,15]
labels = ['Low','Medium','High','Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie(area,labels = labels, colors = colors, startangle = 45)
plt.show()
```

```python
plt.figure(figsize = (9,9)) 
area = [48,30,20,15]
labels = ['Low','Medium','High','Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie(area,labels = labels, colors = colors,shadow = 'true',autopct = '%1.1f%%', startangle = 45)
plt.show()
```

```python
plt.figure(figsize = (9,9)) 
area = [48,30,20,15]
total = np.sum(area)
def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x,total*x/100)

labels = ['Low','Medium','High','Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie(area,labels = labels, colors = colors,shadow = 'true',autopct = val_per, startangle = 45)
plt.show()
```
