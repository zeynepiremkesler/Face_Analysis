```python
import numpy as np
from scipy import linalg
```

```python
A = np.array([[1,2],[3,4]])
print(A)
B = linalg.inv(A)
print(B)
```

```python
A = np.array([[1,2],[3,4]])
print(A)
print(A.T)
```

```python
X = np.array([[1,3,5],[2,5,1],[2,3,8]])
print(X)
y = np.array([[10],[8],[3]])
print(y)
print('Matris tersi ile denklem çözümü \n', linalg.inv(X).dot(y))
print('Solve fonksiyonu ile çözüm \n', linalg.solve(X,y)) # Hızlı yol
```

```python
from scipy import integrate
x2 = lambda x : x**2
I = integrate.quad(x2,0,6)
print(I)
```

```python
from scipy.misc import derivative
def f(x):
    return x**3 + x**2
D = derivative(f,1.0, dx = 1e-6)
print(D)
```
